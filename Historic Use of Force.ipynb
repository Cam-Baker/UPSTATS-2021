```python
df = pd.DataFrame()
for filename in filelist:
    df_0 = pd.read_csv(filename, na_values=np.nan)
    df_0.rename(columns=rename_dict[filename], inplace=True)
    df_0=df_0[col_selection]
    df_0['Date Occurred'] = pd.to_datetime(df_0['Date Occurred'])
    
    #fix race variable and drop the other two 
    df_0['Race'] = df_0['Subject Race'].copy()
    df_0['Race'][(df_0['Subject Ethnicity'] == 'H') & (df_0['Subject Race'] == 'W')] = "H"
    df_0.drop(['Subject Race','Subject Ethnicity'], axis=1, inplace=True)
    
    #fix level of force: mixed force-> highest level of force 
    df_0['R2R Level'] = df_0['R2R Level'].replace({34:3,
                           23:2,
                           234:2,
                           24:2})
    
    ## I commented this out b/c two police beating up one person should count as multiple records. Just recognize what you're measuring. It's not the count of subjects that force was used against. It's instances of force. 
    #drop duplicate RIN's 
    #difference = len(df_0['RIN']) - len(df_0['RIN'].unique())
    #print(f"{filename}: has {difference} duplicates removed.")
    #df_0.drop_duplicates(subset=['RIN'],inplace=True)
    
    #Only look at traffic stop force 
    df_0['Nature of Contact'].replace("1-TRAFFIC STOP [Motor Vehicle Stop]","TRAFFIC STOP", inplace=True)
    df_0 = df_0[df_0['Nature of Contact'] == "TRAFFIC STOP"]
    df_0.drop(['Nature of Contact'],axis=1,inplace=True)
    
    #append to df 
    df = df.append(df_0)

## Good new, RIN's aren't repeated between years. 
#check if RIN's are reused between the years... 
#difference = len(df['RIN']) - len(df['RIN'].unique())
#print(f"total has {difference} duplicates removed.")
```
