```python
def percents(df, colname):
    #returns the results of a search as a percentage of the total searches for that race 
    race_counts = df_arr.groupby(df_arr['APD_RACE_DESC'])[colname].count().to_frame()
    return df_arr.groupby(df_arr['APD_RACE_DESC'])[colname].value_counts().to_frame() / race_counts
```

### In either case... Find force/arrest 

```python
filename = "2019_Response_to_Resistance_Data.csv"
df_rr = pd.read_csv(filename, index_col =0, na_values = np.nan)
column_names19 = {'RIN':'RIN',
        'Primary Key':'Primary Key',
        'Date  Occurred':'Date Occurred',
        'Time  Occurred':'Time Occurred',
        'Location':'Location',
        'Area_Command':'Area Command',
        'Nature of Contact':'Nature of Contact',
        'Reason Desc':'Reason Desc',
        'R2R Level':'R2R Level',
        'Master Subject ID':'Master Subject ID',
        'Subject  Sex':'Subject Sex',
        'Subject  Race':'Subject Race',
        'Subject  Ethnicity':'Subject Ethnicity',
        'Subject Conduct Desc':'Subject Conduct Desc',
        'Subject Resistance':'Subject Resistance',
        'Weapon Used 1':'Weapon Used 1',
        'Weapon Used 2':'Weapon Used 2',
        'Weapon Used 3':'Weapon Used 3',
        'Weapon Used 4':'Weapon Used 4',
        'Weapon Used 5':'Weapon Used 5',
        'Number  Shots':'Number Shots',
        'Effect on Officer':'Effect on Officer',
        'Officer  Organization Desc':'Officer Organization Desc',
        'Officer  Commission Date':'Officer Commission Date',
        'Officer Yrs of Service':'Officer Yrs of Service',
        'X-Coordinate':'X-Coordinate',
        'Council District':'Council District',
        'Census Tract':'Census Tract',
        'ZIP':'ZIP',
        'County':'County',
        'County desc':'County desc'}
RACE_DICT = {'H':'HISPANIC OR LATINO',
            'W':'WHITE',
            'B':'BLACK',
            'A':'ASIAN',
            'M':'MIDDLE EASTERN',
            'U':'UNKNOWN',
            'I':'AMERICAN INDIAN/ALASKAN',
            'P':'HAWAIIAN/PACIFIC ISLAND'}
df_rr.rename(columns=column_names19,inplace=True)
df_rr['Date Occurred'] = pd.to_datetime(df_rr['Date Occurred'])
#fix race variable and drop the other two 
df_rr['Race'] = df_rr['Subject Race'].copy()
df_rr['Race'][(df_rr['Subject Ethnicity'] == 'H') & (df_rr['Subject Race'] == 'W')] = "H"
df_rr.drop(['Subject Race','Subject Ethnicity'], axis=1, inplace=True)
df_rr['Race'] = df_rr['Race'].replace(RACE_DICT)

#fix level of force: mixed force-> highest level of force 
df_rr['R2R Level'] = df_rr['R2R Level'].replace({34:3,
                       23:2,
                       234:2,
                       24:2})
df_rr_mv = df_rr[df_rr['Nature of Contact'] == "1-TRAFFIC STOP [Motor Vehicle Stop]"]
df_rr_mv.head(2)
```

```python
df_rr_mv['Race'].value_counts().to_frame()
```

```python
#Without dropping duplicates
df_fa = df_arr['APD_RACE_DESC'].value_counts().to_frame()
df_fa.rename(columns={'APD_RACE_DESC':'Arrests'}, inplace=True)
df_fa['Force'] = df_rr_mv['Race'].value_counts()
df_fa['F-A'] = df_fa['Force'] / df_fa['Arrests']
df_fa
```
